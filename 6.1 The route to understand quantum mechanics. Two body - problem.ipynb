# Project 6.1: The route to understand quantum mechanics. Two body - problem

Quick dive into two body - problem

## 1. Problem formulation

I would like to start with introducing Lagrangian of our system containing three terms responsible for kinetic energy of first body, kinetic energy of second body and their interaction respectively:

$\mathcal{L}(\dot{q_1}, \dot{q_2}, p_1, p_2) = \frac{m_1 \dot{q_1}^2}{2} + \frac{m_2 \dot{q_2}^2}{2} + G \frac {m_1 m_2}{|q_1 - q_2|}$

To simplify the problem I would like to assume: $m_1 = m_2 = 2$ and $Gm_1 m_2 = 1$ to obtain:

$\mathcal{L}(\dot{q_1}, \dot{q_2}, p_1, p_2) = \dot{q_1}^2+ \dot{q_2}^2 + \frac 1{|q_1 - q_2|}$

## 2. First trick - center of mass

To reduce number of variables I would like to introduce two new variables, namely center of mass:

$$r = \frac{q_1 + q_2}{2}$$

and distance between our bodies:

$$R = q_2 - q_1$$

Substituting into Lagrangian will yield:

$$\mathcal{L} = 2 \dot{r}^2 + \dot{R}^2 + \frac{1}{R} =  2 \dot{r}^2 + \frac{1}{|R|} $$

Since $\dot{R}^2 << 1$

## 3. Second trick - Spherical/Polar coordinate system

Since we would like to solve our problem in 3D I would like to introduce spherical coordinate system, where square of infinitesimal distance equal:

$$dr^2 = \rho^2 d\theta^2 + d\rho^2 + \rho^2sin^2\theta d\phi^2$$

Dividing by $dt^2$ should yield:

$$\dot{r}^2 = \rho^2 \dot{\theta}^2 + \dot{\rho}^2 + \rho^2sin^2\theta \dot{\phi}^2$$

So our Lagrangian looks the following way:

$\mathcal{L} = 2\big(\rho^2 \dot{\theta}^2 + \dot{\rho}^2 + \rho^2sin^2\theta \dot{\phi}^2\big) +  \frac{1}{|R|} $

## 4. Third trick - conservation rules

There is more than one reason to state that angular momentum L is conserved, thus no net moment of force is expected to act on our system, we can assume planar motion with $\phi = const$. This cancels one term containing $\dot{\phi}$

$\mathcal{L} = 2\big(\rho^2 \dot{\theta}^2 + \dot{\rho}^2 \big) + \frac{1}{|R|}$

Since in Lagrangian in not implicitelly dependant on $\theta$, it is possible to calculate canonical momentum $p_\theta$ using cyclic coordinate $\theta$. However $p_\theta$ is just angular momentum we have:

$L =  p_\theta = \frac{\partial \mathcal{L} }{\partial \dot{\theta} } = 4\rho^2 \dot{\theta} $

Another conserved quantity is energy given by equation:

$E = 2\big(\rho^2 \dot{\theta}^2 + \dot{\rho}^2 \big) - \frac{1}{|R|}$

Substituting one equation into another yield:

$ E = 2\big(\frac{L^2}{4\rho^2} + \dot{\rho}^2 \big) - \frac{1}{|R|} = 2\dot{\rho}^2 + U_{eff}$

Where we defined effective energy $U_{eff} = L^2/4\rho^2 - \frac{1}{|R|}$, which effectively is a constant depending on initial conditions.

Finally:

$$\dot{\rho} = \sqrt{\frac{E - U_{eff}}{2} }$$

Using one more time angular momentum equation:

$$\dot{\theta} = \frac{L^2}{4\rho^2}$$

## 5. Final step and integrating:

Finally we need to divide one equation by another to find trajectory:

$$ d\theta = \frac{1}{2} \frac{L^2}{\rho^2(E-U_{eff})} d\rho$$  

$$\theta(t) -  \theta(0) = \int_{r_0}^{r}\frac{L^2}{\rho^2 \sqrt{2(E+\frac{1}{|R|})-L^2/4\rho^2}} d\rho=  \int_{r_0}^{r}\frac{L^2}{\rho^2 \sqrt{a} \sqrt{1-\frac{L^2}{4a\rho^2}}}d\rho$$

Substituting $u = \frac{L}{2\sqrt{a}\rho}$ we have:

$$\theta(t) -  \theta(0) = 2L \int_{L/2r_0\sqrt{a}}^{L/2r\sqrt{a}} \frac{-du}{\sqrt{1-u^2}} = L \ arccos(u) \bigg|_{L/2r_0\sqrt{a}}^{L/2r\sqrt{a}} = 2L \ arccos(L/2r\sqrt{a}) -  2L \ arccos(L/2r_0\sqrt{a}) = 2L \ arccos(L/2r\sqrt{a}) -LC_0 $$

setting $ \ \theta(0) = 0$

### $cos\big(\theta(t)/2L+ C_0\big) = \frac{L}{2r} \frac{1}{\sqrt{2(E+1/|R|)}}$

$$r = \frac{L}{2\sqrt{2(E+1/|R|)}} \frac{1}{cos\big(\theta(t)/2L+ C_0\big)}$$

## 6. Program

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
times = np.linspace(0,1000,1001)
phi0 = 0 
r0 = 100
L = 1000
R = 100
E = 0
r = []
phis = np.linspace(0,16*np.pi,801)
C0 = np.arccos(L/2/r0*np.sqrt(2*(E+1/R)))
print(L/2/r0*np.sqrt(2*(E+1/R)))
print(C0)

for phi in phis:
    r.append(2*L/(np.cos(phi/L + C0)*2*np.sqrt(2*(E+1/R))))
    print('Eff energy>:')
    print(L**2/4/r[-1]**2 - 1/R)
    print('radius:')
    print(r[-1])
```

```python
fig = plt.figure(facecolor='white')
ax = plt.subplot(111, polar=True)
ax.plot(phis, r)
ax.set_rmax(20000)
plt.show()
```
